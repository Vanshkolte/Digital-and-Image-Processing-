```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def segment_image(image_path):
    # Load the image
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"Error: Could not read image at {image_path}")
        return

    # i) Horizontal Lines Detection
    horizontal_kernel = np.ones((2, 19), np.uint8)  # Adjust kernel size as needed
    horizontal_lines = cv2.erode(img, horizontal_kernel, iterations=1)

    plt.subplot(231),plt.imshow(img, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(232),plt.imshow(horizontal_lines, cmap='gray')
    plt.title('Horizontal Lines'), plt.xticks([]), plt.yticks([])

    plt.show()

# Example usage
image_path = "/content/lines (1).png"  # Replace with your image path
segment_image(image_path)
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def segment_image(image_path):
    # Load the image
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"Error: Could not read image at {image_path}")
        return

    # iii) Canny Edge Detector
    edges = cv2.Canny(img, 50, 150) # Adjust thresholds as needed

    # iv) Hough Line Transform
    lines = cv2.HoughLines(edges, 1, np.pi / 180, 200)  # Adjust threshold as needed
    hough_lines_image = img.copy()
    if lines is not None:
        for line in lines:
            rho, theta = line[0]
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a * rho
            y0 = b * rho
            x1 = int(x0 + 1000 * (-b))
            y1 = int(y0 + 1000 * (a))
            x2 = int(x0 - 1000 * (-b))
            y2 = int(y0 - 1000 * (a))
            cv2.line(hough_lines_image, (x1, y1), (x2, y2), (255, 0, 0), 2)

    # Display the results
    plt.figure(figsize=(20, 15))

    plt.subplot(231),plt.imshow(img, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(232),plt.imshow(hough_lines_image, cmap='gray')
    plt.title('Hough Lines'), plt.xticks([]), plt.yticks([])

    plt.show()

# Example usage
image_path = "/content/parking_lot (1).jpg"  # Replace with your image path
segment_image(image_path)
```
