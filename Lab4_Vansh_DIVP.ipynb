```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
from google.colab.patches import cv2_imshow
```

```python
# 1] Plot a Histogram of grayscale and Coloring
# Load the image in grayscale
gray_image = cv2.imread('/content/Lenacolor.jpg', cv2.IMREAD_GRAYSCALE)
cv2_imshow(gray_image)
# 1) Plot Histograms of Grayscale and Color Image
def plot_histograms():
    # Grayscale histogram
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.hist(gray_image.ravel(), bins=256, color='grey')
    plt.title("Grayscale Histogram")
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")
plot_histograms()
```

```python
# Load the image in color
color_image = cv2.imread('/content/Lenacolor.jpg')
cv2_imshow(color_image)

# Color histograms
plt.figure(figsize=(8, 5))

# Compute and sum the histograms for all three channels
hist_r = cv2.calcHist([color_image], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([color_image], [1], None, [256], [0, 256])
hist_b = cv2.calcHist([color_image], [2], None, [256], [0, 256])

# Sum the histograms
combined_hist = hist_r + hist_g + hist_b

plt.plot(combined_hist, color='blue')
plt.title("Combined RGB Histogram (Blue)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()


# Rgb color
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
for i, color in enumerate(['r', 'g', 'b']):
        hist = cv2.calcHist([color_image], [i], None, [256], [0, 256])
        plt.plot(hist, color=color)
plt.title("Color Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# 3] Perfrom Histogram Equalisation on gray Scale

import cv2
import matplotlib.pyplot as plt

# Load the grayscale image
gray_image = cv2.imread('/content/Lenacolor.jpg', cv2.IMREAD_GRAYSCALE)

# Perform Histogram Equalization
equalized_image = cv2.equalizeHist(gray_image)

# Plot Original and Equalized Images with Histograms
plt.figure(figsize=(12, 8))

# Original Image and Histogram
plt.subplot(2, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis('off')

plt.subplot(2, 2, 2)
plt.hist(gray_image.flatten(), bins=5, color='lightblue')
plt.title("Original Grayscale Histogram (bins=5)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

# Equalized Image and Histogram
plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis('off')

plt.subplot(2, 2, 4)
plt.hist(equalized_image.flatten(), bins=5, color='lightblue')
plt.title("Equalized Grayscale Histogram (bins=5)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
