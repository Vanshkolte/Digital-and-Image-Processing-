```python
# AND of two images
image1 = cv2.imread('/content/drive/MyDrive/ImageProcesing/Lab2/Logical1.jpg')
image2 = cv2.imread('/content/drive/MyDrive/ImageProcesing/Lab2/logical2.jpg')

 ## (ensure both images have the same size)
image1 = cv2.resize(image1, (180, 180))
image2 = cv2.resize(image2, (180, 180))

print('Image 1')
cv2_imshow(image1)
print('Image 2')
cv2_imshow(image2)

# bitwise AND
print('bitwise_and')
bitwise_and = cv2.bitwise_and(image2, image1)
cv2_imshow(bitwise_and)
```

```python
# Bitwise OR
print('Image 1')
cv2_imshow(image1)
print('Image 2')
cv2_imshow(image2)
print('bitwise_or')
bitwise_or = cv2.bitwise_or(image2, image1)
cv2_imshow(bitwise_or)
```

```python
# Bitwise XOR
print('Image 1')
cv2_imshow(image1)
print('Image 2')
cv2_imshow(image2)
print('bitwise_xor')
bitwise_xor = cv2.bitwise_xor(image2, image1)
cv2_imshow(bitwise_xor)
```

```python
# bitwise NOT
image1 = cv2.imread('/content/drive/MyDrive/ImageProcesing/Lab2/NOT operation.jpg')
print('Image 1')
cv2_imshow(image1)
print('bitwise_not')
bitwise_not = cv2.bitwise_not(image1)
cv2_imshow(bitwise_not)
```
