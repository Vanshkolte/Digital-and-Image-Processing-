```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the images
img0 = cv2.imread('/content/images O.png', cv2.IMREAD_GRAYSCALE)
imgH = cv2.imread('/content/Letter H.jpg', cv2.IMREAD_GRAYSCALE)

# Define the structuring element (kernel)
kernel = np.ones((5,5), np.uint8)

# Step 1: Perform Dilation for 'O'
print("Original Image ")
cv2_imshow(img0)
```

```python

print("Performing Dilation on Image 0...")
dilation0 = cv2.dilate(img0, kernel, iterations=4)
cv2_imshow(dilation0)  # Display dilation result
```

```python

# Step 2: Perform Erosion for 'O'
print("Performing Erosion on Image 0...")
erosion0 = cv2.erode(img0, kernel, iterations=10)
cv2_imshow(erosion0)  # Display erosion result
```

```python

# Step 3: Perform Opening for 'O' (Erosion followed by Dilation)
print("Performing Opening on Image 0...")
opening0 = cv2.morphologyEx(img0, cv2.MORPH_OPEN, kernel, iterations=10)
cv2_imshow(opening0)  # Display opening result

```

```python
# Step 4: Perform Closing for 'O' (Dilation followed by Erosion)
print("Performing Closing on Image 0...")
closing0 = cv2.morphologyEx(img0, cv2.MORPH_CLOSE, kernel,iterations=6)
cv2_imshow(closing0)  # Display closing result

```

```python
# Step 1: Perform Dilation for 'H'
print("Performing Dilation on Image H...")
dilationH = cv2.dilate(imgH, kernel, iterations=4)
cv2_imshow(dilationH)  # Display dilation result

# Step 2: Perform Erosion for 'H'
print("Performing Erosion on Image H...")
erosionH = cv2.erode(imgH, kernel, iterations=10)
cv2_imshow(erosionH)  # Display erosion result

# Step 3: Perform Opening for 'H' (Erosion followed by Dilation)
print("Performing Opening on Image H...")
openingH = cv2.morphologyEx(imgH, cv2.MORPH_OPEN, kernel)
cv2_imshow(openingH)  # Display opening result

# Step 4: Perform Closing for 'H' (Dilation followed by Erosion)
print("Performing Closing on Image H...")
closingH = cv2.morphologyEx(imgH, cv2.MORPH_CLOSE, kernel)
cv2_imshow(closingH)  # Display closing result
```
